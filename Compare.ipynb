```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
sim_dic = {}
seq_dic = {}
lr_list = ['0.01','0.003','0.001','0.0003']

mode = 'train'
for lr in lr_list:
    if lr=='0.001':
        with open('saves/base_model/log_%s.txt'%mode) as f:
            lines = f.read().strip().split('\n')
        lines = lines[:len(seq_dic[lr_list[0]])]
    else:
        file='saves/base_lr_%s/log_%s.txt'%(lr,mode)
        with open(file) as f:
            lines = f.read().strip().split('\n')
    seq_dic[lr]=[]
    sim_dic[lr]=[]
    for line in lines:
        seq_dic[lr].append(line.split(':')[-1].strip())
        sim_dic[lr].append(line.split(':')[-2].split('\t')[0].strip())
    x = np.arange(len(seq_dic[lr]))
    y = np.array([float(x) for x in seq_dic[lr]])
    plt.plot(x,y,label=lr)
plt.legend(loc='best')
plt.show()
```
